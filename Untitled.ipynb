```python
import re

data = """
Jay Harmishton      jay.harmishton@yu.DAV.edu
Adele D. Helmsley-Chintsworth    adele.d.chintsworth@yu.AIM.edu
Renfrew Thoreau Donnington Swill     renfrew.swll@yu.QE.edu
Kanafra Mustafa Soapdi  kanafra.m.soapdi@yu.MAT.edu
Xiaoli Kzhang       xiaoli.kzhang@yu.CYB.edu
"""

pattern = r"(\b[A-Za-z]+\b\.\b[A-Za-z]+\b)@(\b[A-Za-z]+\b\.\b[A-Za-z]+\b)"
matches = re.findall(pattern, data)

for match in matches:
    f_name, l_name = match[0].split('.')
    domain = match[1]
    print(f"{f_name.capitalize()} {l_name.capitalize()} {domain}")
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
# Forward fill to handle NaN values in 'Period' and 'Unnamed: 1' columns
df['Period'] = df['Period'].fillna(method='ffill')
df['Unnamed: 1'] = df['Unnamed: 1'].fillna('Exports')

# Melt the dataframe to long format
df_melted = pd.melt(df, id_vars=['Period', 'Unnamed: 1'])

# Rename the columns
df_melted.columns = ['Month', 'Type', 'Year', 'Value']

# Remove rows with NaN values in 'Value' column
df_melted.dropna(subset=['Value'], inplace=True)

# Sort values by Year and reset index
df_melted = df_melted.sort_values(by=['Year', 'Month']).reset_index(drop=True)

df_melted.head(20)
```
